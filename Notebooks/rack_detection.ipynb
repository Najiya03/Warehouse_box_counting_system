**Rack detection from front view**

```python
import os

dataset_path = "/content/drive/MyDrive/Warehouse_box_counting/datasets/Front_rack_dataset"

for root, dirs, files in os.walk(dataset_path):
    print(root)
```

```python
yaml_content = f"""
path: {dataset_path}

train: images/train
val: images/val

names:
  0: rack
"""

with open("/content/data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created")
```

```python
with open("/content/data.yaml", "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    project="/content/drive/MyDrive/Warehouse_box_counting/outputs",
    name="front_rack_yolov8n"
)
```

```python
import shutil

shutil.copy(
    "/content/drive/MyDrive/Warehouse_box_counting/outputs/front_rack_yolov8n/weights/best.pt",
    "/content/drive/MyDrive/Warehouse_box_counting/models/front_rack_best.pt"
)

print("Model saved successfully")
```

```python
from ultralytics import YOLO

model = YOLO(
    "/content/drive/MyDrive/Warehouse_box_counting/models/front_rack_best.pt"
)

model.predict(
    source="/content/drive/MyDrive/Warehouse_box_counting/datasets/Front_rack_dataset/images/val",
    save=True,
    conf=0.6,
    project="/content/drive/MyDrive/Warehouse_box_counting/outputs",
    name="front_validation_predictions"
)
```

```python
model.predict(
    source="/content/drive/MyDrive/WarehouseBox/B93_Front.mp4",
    conf=0.6,
    save=True,
    verbose=True,
    project="/content/drive/MyDrive/Warehouse_Box_Counting/outputs",
    name="B93_front_detection"
)
```

```python
from ultralytics import YOLO

model = YOLO(
    "/content/drive/MyDrive/Warehouse_box_counting/models/front_rack_best.pt"
)
```

```python
results = model.predict(
    source="/content/drive/MyDrive/Warehouse_box_counting/datasets/Front_test.mp4",
    conf=0.6,
    save=True,
    stream=True,
    project="/content/drive/MyDrive/Warehouse_Box_Counting/outputs",
    name="Front_view_rack_detection",
    exist_ok=True
)

# Consume the generator so the whole video is processed
for _ in results:
    pass

print("Video processing completed.")
```

**Rack detection from top view**

```python
from ultralytics import YOLO
from PIL import Image
from IPython.display import display
import os

model = YOLO(
    "/content/drive/MyDrive/Warehouse_box_counting/models/front_rack_best.pt"
)

folder = "/content/drive/MyDrive/Warehouse_box_counting/datasets/topViewSamples"

for img in os.listdir(folder):

    if img.endswith((".jpg", ".png", ".jpeg")):

        print("\nTesting:", img)

        results = model.predict(
            source=os.path.join(folder, img),
            conf=0.5
        )

        display(Image.fromarray(results[0].plot()))

        print("Number of detections:", len(results[0].boxes))
```
