**Rack detection from front view**

```python
import os

dataset_path = "/content/drive/MyDrive/Warehouse_box_counting/datasets/Front_rack_dataset"

for root, dirs, files in os.walk(dataset_path):
    print(root)
```

```python
yaml_content = f"""
path: {dataset_path}

train: images/train
val: images/val

names:
  0: rack
"""

with open("/content/data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created")
```

```python
with open("/content/data.yaml", "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    project="/content/drive/MyDrive/Warehouse_box_counting/outputs",
    name="front_rack_yolov8n"
)
```

```python
import cv2

input_video = "/content/drive/MyDrive/WarehouseBox/B93_Front.mp4"
output_video = "/content/drive/MyDrive/WarehouseBox/B93_Front_720p.mp4"

cap = cv2.VideoCapture(input_video)

fps = cap.get(cv2.CAP_PROP_FPS)

width = 1280
height = 720

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (width, height))
    out.write(frame)

    frame_count += 1

    if frame_count % 1000 == 0:
        print(f"{frame_count} frames processed")

cap.release()
out.release()

print("Video resized successfully")
```

**Rack detection from top view**

```python
from ultralytics import YOLO
from PIL import Image
from IPython.display import display
import os

model = YOLO(
    "/content/drive/MyDrive/Warehouse_box_counting/models/front_rack_best.pt"
)

folder = "/content/drive/MyDrive/Warehouse_box_counting/datasets/topViewSamples"

for img in os.listdir(folder):

    if img.endswith((".jpg", ".png", ".jpeg")):

        print("\nTesting:", img)

        results = model.predict(
            source=os.path.join(folder, img),
            conf=0.5
        )

        display(Image.fromarray(results[0].plot()))

        print("Number of detections:", len(results[0].boxes))
```
